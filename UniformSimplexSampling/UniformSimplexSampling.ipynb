```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def simplex_sampler(n,M):
    c_side = [i for i in range(10**4)]
    r = list(np.random.choice(c_side, size=n-1, replace=False, p=None))
    r = [0]+r+[10**4]
    rd = [(r[i]-r[i-1])/10**4 for i in range(1,len(r))]
    return rd
```

```python
simplex_sampler(3,10**4)
```

```python
def random_portfolios(n,a):
    portfolios = [simplex_sampler(n,10**4) for i in range(a)]
    return portfolios
```

```python
st = random_portfolios(3,100000)
```

```python
x = [st[i][1] for i in range(100000)]
y = [st[i][2] for i in range(100000)]
```

```python
plt.figure(figsize=(15,15)) 
sns.scatterplot(x,y)
```
